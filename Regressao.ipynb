```python
import kagglehub
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
from sklearn.linear_model import Lasso, Ridge
```

```python
# Download latest version
path = kagglehub.dataset_download("muhamedumarjamil/crypto-and-gold-prices-dataset-20152025")
# Lista dos arquivos csv do download
lista_arquivos = os.listdir(path)

try:
  df = pd.read_csv(os.path.join(path, lista_arquivos[0]))
  print('Dados carregados com sucesso')
except Exception as e:
  print(f'Erro ao carregar dados: {e}')
  df = None
```

```python
X = df.drop(['Date', 'Gold (USD per oz)'], axis=1)
y = df['Gold (USD per oz)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Regressão linear
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)

# Medidas de avaliação
# Mean Absolute Error: trata todos os erros igualmente
lr_mae = mean_absolute_error(y_test, lr_pred)
print(f'MAE: {lr_mae}')

# Mean Squared Error: penaliza os erros maiores
lr_mse = mean_squared_error(y_test, lr_pred)
print(f'MSE: {lr_mse}')

# Root Mean Squared Error: sensível aos grandes erros
lr_rmse = root_mean_squared_error(y_test, lr_pred)
print(f'RMSE: {lr_rmse}')

# R-squared: indica a proporção da variabilidade dos dados explicada pelo modelo
lr_r2 = r2_score(y_test, lr_pred)
print(f'R2: {lr_r2}')
```

```python
# Métodos de regularização
# Lasso
lasso = Lasso(alpha=2)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)

lasso_mae = mean_absolute_error(y_test, lasso_pred)
lasso_mse = mean_squared_error(y_test, lasso_pred)
lasso_rmse = root_mean_squared_error(y_test, lasso_pred)
lasso_r2 = r2_score(y_test, lasso_pred)
print('Lasso')
print(f'MAE: {lasso_mae}')
print(f'MSE: {lasso_mse}')
print(f'RMSE: {lasso_rmse}')
print(f'R2: {lasso_r2}')

# Ridge
ridge = Ridge(alpha=2)
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)

ridge_mae = mean_absolute_error(y_test, ridge_pred)
ridge_mse = mean_squared_error(y_test, ridge_pred)
ridge_rmse = root_mean_squared_error(y_test, ridge_pred)
ridge_r2 = r2_score(y_test, ridge_pred)
print('\nRidge')
print(f'MAE: {ridge_mae}')
print(f'MSE: {ridge_mse}')
print(f'RMSE: {ridge_rmse}')
print(f'R2: {ridge_r2}')
```
